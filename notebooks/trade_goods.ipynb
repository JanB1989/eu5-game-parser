## Merged Vanilla + Prosper or Perish

This uses `eu5_load_order.toml` and the `merged_default` profile, so vanilla is loaded first and Prosper or Perish is applied on top.

```python
from eu5gameparser import load_eu5_data

merged = load_eu5_data(profile="merged_default")
merged.goods.filter(merged.goods["name"].is_in(["victuals", "wheat", "fish", "livestock"]))
```

```python
from eu5gameparser.graphs import show_good_flow

show_good_flow("victuals", depth=2, data=merged.building_data, enable_zoom=True)
```
