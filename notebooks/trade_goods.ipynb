## Merged Vanilla + Mods

This uses `eu5_load_order.toml` and the `merged_default` profile, so vanilla is loaded first and configured mods are applied on top.


```python
from eu5gameparser import load_eu5_data

merged = load_eu5_data(profile="merged_default")
merged.goods_summary.head()
```

```python
import polars as pl

raw_material_output_counts = (
    merged.goods_summary
    .filter(pl.col("category") == "raw_material")
    .select(["name", "output_method_count", "input_method_count", "default_market_price"])
    .sort(["output_method_count", "name"])
)

with pl.Config(tbl_rows=-1):
    display(raw_material_output_counts)
```

## Goods-Flow Explorer

Update and open the single standalone explorer. Select the good, max age, specific unlocks, and depth inside the popout.


```python
from eu5gameparser.graphs import open_goods_flow_explorer

open_goods_flow_explorer(
    path="out/goods_flow_explorer.html",
    eu5_data=merged,
)
```
